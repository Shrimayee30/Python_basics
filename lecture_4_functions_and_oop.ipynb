## 1. Why Do We Need Functions?

Sometimes we write the same code again and again.

A **function** helps us group code into one reusable block. Instead of rewriting the same steps, we can call the function whenever we need it.

````python
print("Welcome, Asha!")
print("Welcome, Asha!")
print("Welcome, Asha!")
````

This works, but it repeats the same idea. Functions give us a cleaner way.

````python
def welcome():
    print("Welcome, Asha!")

welcome()
welcome()
welcome()
````

## 2. What Is a Function?

A function is a named block of code that performs a task.

Basic idea:

```python
def function_name():
    do something
```

To use the function, we call it with its name.

## 3. First Function Example

````python
def say_hello():
    print("Hello!")

say_hello()
````

Here:

- `def` starts the function definition
- `say_hello` is the function name
- the indented code is the body of the function
- `say_hello()` runs the function

## 4. Functions with Parameters

A parameter lets a function receive information.

This makes the function more flexible.

````python
def greet(name):
    print("Hello,", name)

greet("Riya")
greet("Kabir")
````

In this example:

- `name` is the parameter
- `"Riya"` and `"Kabir"` are arguments passed to the function

## 5. Functions with More Than One Parameter

````python
def add_numbers(a, b):
    print(a + b)

add_numbers(4, 5)
add_numbers(10, 3)
````

## 6. `return` in Functions

Some functions only print something.
Some functions calculate a value and give it back using `return`.

````python
def multiply(a, b):
    return a * b

result = multiply(3, 4)
print(result)
````

Why `return` is useful:

- we can store the answer in a variable
- we can use the answer later
- the function becomes more powerful

## 7. `print` vs `return`

Students often confuse these two ideas.

- `print()` shows something on the screen
- `return` sends a value back from the function

````python
def show_square(n):
    print(n * n)

def give_square(n):
    return n * n

show_square(5)
answer = give_square(5)
print(answer)
````

## 8. Worked Table: How a Function Call Updates Values

Let us trace this example:

```python
def multiply(a, b):
    return a * b

result = multiply(3, 4)
```

| Step | What happens | Value |
|---|---|---|
| 1 | Function is called | `multiply(3, 4)` |
| 2 | `a` receives the first argument | `a = 3` |
| 3 | `b` receives the second argument | `b = 4` |
| 4 | Expression is calculated | `a * b = 12` |
| 5 | Function returns the value | `12` |
| 6 | Returned value is stored | `result = 12` |

## 9. Small Function Examples

````python
def say_good_morning():
    print("Good morning!")

def double_number(x):
    return x * 2

say_good_morning()
print(double_number(9))
````

## 11. Transition: From Functions to OOP

Functions are great when we want to organize actions.

But sometimes we want to represent **things** in code, not just actions.

For example:

- a student
- a car
- a bank account
- a book

This is where OOP becomes useful.

## 12. What Is Object-Oriented Programming?

Object-oriented programming, or **OOP**, is a way of writing programs by thinking about real-world objects.

In OOP, we model things using:

- **class**: a blueprint
- **object**: an actual item made from the blueprint
- **attributes**: the data of the object
- **methods**: the actions of the object

## 13. Real-Life Analogy

Think about a `Car`.

- The **class** is the design or blueprint of a car.
- An **object** is one actual car.
- Its **attributes** could be color, brand, and speed.
- Its **methods** could be start, stop, and drive.

## 14. First Class Example

In Python, we create a class using the `class` keyword.

````python
class Student:
    pass
````

This creates a class named `Student`.

Right now it does nothing, but it gives us the basic idea of a blueprint.

## 15. Creating an Object

````python
class Student:
    pass

s1 = Student()
print(s1)
````

Here:

- `Student` is the class
- `s1` is an object of that class

## 16. Attributes

Attributes are pieces of data that belong to an object.

````python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

s1 = Student("Aarav", 8)
print(s1.name)
print(s1.grade)
````

Important ideas:

- `__init__` runs when the object is created
- `self` means the current object
- `self.name` and `self.grade` are attributes

## 17. Methods

Methods are functions written inside a class.

They describe what an object can do.

````python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def introduce(self):
        print("My name is", self.name)
        print("I am in grade", self.grade)

s1 = Student("Aarav", 8)
s1.introduce()
````

This is a nice OOP idea:

- data and behavior are connected
- the object stores information
- the method uses that information

## 18. Worked Table: Object Creation and Attribute Update

Let us trace this line:

```python
s1 = Student("Aarav", 8)
```

| Step | What happens | Value |
|---|---|---|
| 1 | Class is called | `Student("Aarav", 8)` |
| 2 | A new object is created | `s1` object |
| 3 | `name` receives the first value | `name = "Aarav"` |
| 4 | `grade` receives the second value | `grade = 8` |
| 5 | `self.name` is stored in the object | `"Aarav"` |
| 6 | `self.grade` is stored in the object | `8` |
| 7 | The object is assigned to a variable | `s1` |

## 19. Function vs Method

This is a helpful comparison:

- a **function** is written outside a class
- a **method** is a function inside a class

Example:

- `greet("Riya")` is a function call
- `s1.introduce()` is a method call

## 20. Why Learn OOP?

OOP helps us organize code when programs become bigger.

It is especially useful when we want to model real-world things clearly.
