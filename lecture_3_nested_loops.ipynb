## 1. Quick Review: One Loop

Before we learn nested loops, let us remember what a single loop does.

A loop repeats a block of code.

````python
for i in range(3):
    print("Hello")
````

### Worked Table: Example 1

When students are learning loops, it helps to trace each iteration.

| Iteration | `i` value | What gets printed |
|---|---:|---|
| 1 | 0 | `Hello` |
| 2 | 1 | `Hello` |
| 3 | 2 | `Hello` |

Notice:

- `range(3)` gives `0, 1, 2`
- the loop runs 3 times
- the printed text stays the same, but the loop variable changes

That loop repeats one action 3 times.

Now let us think bigger: what if we want to repeat a loop itself?

## 2. What Is a Nested Loop?

A nested loop means **a loop inside another loop**.

Basic idea:

```python
for outer in range(...):
    for inner in range(...):
        do something
```

The outer loop controls how many times the inner loop starts.
The inner loop runs completely for each turn of the outer loop.

## 3. First Nested Loop Example

Let us print pairs of numbers.

````python
for row in range(1, 4):
    for col in range(1, 3):
        print("row =", row, ", col =", col)
````

### Worked Table: Example 2

Here we track both loop variables.

| Step | `row` value | `col` value | What gets printed |
|---|---:|---:|---|
| 1 | 1 | 1 | `row = 1 , col = 1` |
| 2 | 1 | 2 | `row = 1 , col = 2` |
| 3 | 2 | 1 | `row = 2 , col = 1` |
| 4 | 2 | 2 | `row = 2 , col = 2` |
| 5 | 3 | 1 | `row = 3 , col = 1` |
| 6 | 3 | 2 | `row = 3 , col = 2` |

Notice:

- `row` stays the same while the inner loop is running
- `col` changes faster because it belongs to the inner loop
- when `col` finishes, it starts again from the beginning for the next row

What happens here?

- the outer loop changes `row`
- for each `row`, the inner loop runs again from the beginning
- so every row is matched with every column

## 4. Seeing It More Clearly

This example shows when the outer loop starts and when the inner loop runs.

````python
for row in range(1, 4):
    print("Start row", row)
    for col in range(1, 4):
        print("  col", col)
    print("End row", row)
````

### Worked Table: Example 3

This trace table shows the exact order of events.

| Step | `row` value | `col` value | Action |
|---|---:|---:|---|
| 1 | 1 | - | `Start row 1` |
| 2 | 1 | 1 | `col 1` |
| 3 | 1 | 2 | `col 2` |
| 4 | 1 | 3 | `col 3` |
| 5 | 1 | - | `End row 1` |
| 6 | 2 | - | `Start row 2` |
| 7 | 2 | 1 | `col 1` |
| 8 | 2 | 2 | `col 2` |
| 9 | 2 | 3 | `col 3` |
| 10 | 2 | - | `End row 2` |
| 11 | 3 | - | `Start row 3` |
| 12 | 3 | 1 | `col 1` |
| 13 | 3 | 2 | `col 2` |
| 14 | 3 | 3 | `col 3` |
| 15 | 3 | - | `End row 3` |

Notice:

- the outer loop starts a row
- the inner loop completes all column steps for that row
- only then does the outer loop move to the next row

This helps students see that the inner loop finishes all its turns before the outer loop moves on.

## 5. Rows and Columns

Nested loops are very useful when we work with rows and columns.

Examples:

- classroom seating
- game boards
- tables
- pattern printing

````python
for row in range(1, 4):
    for col in range(1, 5):
        print("*", end=" ")
    print()
````

In this code:

- the inner loop prints stars in one row
- the `print()` after the inner loop moves to the next line
- the outer loop decides how many rows we get

This is one of the easiest nested loop patterns.

````python
for row in range(3):
    for col in range(5):
        print("#", end=" ")
    print()
````

## 7. A Number Table

Nested loops can also print number structures.

````python
for row in range(1, 4):
    for col in range(1, 5):
        print(col, end=" ")
    print()
````

## 8. Multiplication Table Idea

Nested loops are perfect for multiplication tables.

````python
for row in range(1, 4):
    for col in range(1, 6):
        print(row * col, end="\t")
    print()
````

## 11. Pattern Example: Counting Across

````python
for row in range(3):
    for col in range(1, 6):
        print(col, end=" ")
    print()
````

````python
for row in range(3):
    for col in range(5, 0, -1):
        print(col, end=" ")
    print()
````

## 13. Important Pattern Rule

For most beginner patterns:

- outer loop controls rows
- inner loop controls what appears in each row
- a blank `print()` moves to the next line

### Extra Glimpse: 3-Layer and 4-Layer Nested Loops

Most beginner problems use 2 nested loops.

But sometimes we can go deeper:

- 3 layers might mean `floor -> row -> column`
- 4 layers might mean `building -> floor -> row -> column`

You do not need to master 4-layer loops right now. This is just to help students see that the same idea can continue.

````python
for floor in range(1, 3):
    print("Floor", floor)
    for row in range(1, 3):
        for col in range(1, 4):
            print("(", floor, ",", row, ",", col, ")", end=" ")
        print()
    print()
````

In the 3-layer example above:

- the first loop chooses the floor
- the second loop chooses the row
- the third loop chooses the column

So each printed position has 3 values: floor, row, and column.

````python
for building in range(1, 3):
    print("Building", building)
    for floor in range(1, 3):
        for row in range(1, 3):
            for col in range(1, 3):
                print("(", building, ",", floor, ",", row, ",", col, ")", end=" ")
            print()
        print()
````

In the 4-layer example above:

- `building` is the outermost loop
- inside each building, we move through floors
- inside each floor, we move through rows
- inside each row, we move through columns

It is the same loop idea, just with more levels.

## 18. Looking Ahead: Matrices in Math

Nested loops connect very naturally to an important math idea: **matrices**.

A matrix is a rectangular arrangement of numbers written in **rows** and **columns**.

For example, this is a `2 x 3` matrix:

```\n1 2 3\n4 5 6\n```

This means:

- 2 rows
- 3 columns
- each value has a position

We often describe matrix positions using row number and column number.

For the matrix above:

- `1` is in row 1, column 1
- `3` is in row 1, column 3
- `5` is in row 2, column 2
- `6` is in row 2, column 3

## 19. Why Matrices Match Nested Loops

Think about what nested loops already do:

- the outer loop can move through rows
- the inner loop can move through columns
- together, they can visit every position in a table-like structure

That is exactly why nested loops are so useful for matrices.

If we want to read or print a matrix row by row, nested loops are a natural fit.

Math idea:

```\nrow 1 -> col 1, col 2, col 3\nrow 2 -> col 1, col 2, col 3\n```

This is the same thinking we used in pattern printing.
